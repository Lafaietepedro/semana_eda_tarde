```python
import pandas as pd
```

```python
pesquisa = pd.read_csv('datasets/pesquisa_aula_estatistica.csv')
```

## Renomeie todas as colunas

### ['data_hora','email','pontuacao',compreensao','desafiador','sentimento']

```python
pesquisa.columns = ['data_hora','email','pontuacao','compreensao','desafiador','sentimento']
```

```python
pesquisa.head(25)
```

```python
import seaborn as sns
import pandas as pd

# Criar o DataFrame com a contagem de cada sentimento
df = pesquisa['sentimento'].value_counts(normalize=True).reset_index()
df.columns = ['sentimento','Porcentagem']

# Criar o gráfico de barras
ax = sns.barplot(x='sentimento', y='Porcentagem', data=df)

# Adicionar as porcentagens nas barras
ax.bar_label(ax.containers[0], fmt='%.2f%%')

# Mostrar o gráfico
sns.boxplot()
```
